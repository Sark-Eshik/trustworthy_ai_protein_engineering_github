# Combined Sparsity Framework Demonstration

This notebook demonstrates how to orchestrate, merge, and calculate the unified Combined Sparsity metric across empirical, evolutionary, and structural dimensions.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

from src.sparsity.combined.combine_sparsity import CombinedSparsity
from src.sparsity.combined.combined_validation import CombinedValidation
```

## 1. Execute Combined Sparsity Pipeline (Megascale-D Mode)

Megascale-D Mode mathematically integrates three dimensions (Empirical, Evolutionary, and Structural) by taking their average score.

```python
pipeline = CombinedSparsity()
p_path, r_path, plot_path = pipeline.run_pipeline(
    mode="megascale_d",
    empirical_path_override="results/empirical/empirical_sparsity.parquet",
    evolutionary_path_override="results/evolutionary/evolutionary_sparsity.parquet",
    structural_path_override="results/structural/structural_sparsity.parquet",
    output_dir="results/combined/"
)

print(f"Combined Parquet path:      {p_path}")
print(f"Certification Report path:  {r_path}")
print(f"Distribution Plot path:     {plot_path}")
```

## 2. Inspect Combined Results

```python
df = pd.read_parquet(p_path)
df
```

## 3. Run Scientific Math Audit & Certification

The validator runs structural checks and audits the arithmetic calculations for every single mutation record to ensure mathematical correctness.

```python
validator = CombinedValidation()
certified = validator.validate_dataset(p_path, mode="megascale_d")
print(f"Combined Sparsity Framework certified: {certified}")
```

## 4. Run Combined Sparsity Pipeline (TKT Industrial Mode)

In applied enzyme design pipelines (TKT Mode) where empirical frequencies are absent, the framework integrates two dimensions (Evolutionary and Structural) by taking their average score.

```python
tkt_p_path, tkt_r_path, tkt_plot_path = pipeline.run_pipeline(
    mode="tkt",
    empirical_path_override=None,
    evolutionary_path_override="results/evolutionary/evolutionary_sparsity.parquet",
    structural_path_override="results/structural/structural_sparsity.parquet",
    output_dir="results/combined_tkt/"
)

tkt_df = pd.read_parquet(tkt_p_path)
tkt_df
```

## 5. Verify TKT Certification

```python
tkt_certified = validator.validate_dataset(tkt_p_path, mode="tkt")
print(f"TKT Sparsity Framework certified: {tkt_certified}")
```
