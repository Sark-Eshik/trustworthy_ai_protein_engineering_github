# Empirical Sparsity Demonstration

This notebook demonstrates how to load, calculate, and validate empirical sparsity using the Trustworthy AI Protein Engineering pipeline.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

from src.sparsity.empirical.empirical_sparsity import EmpiricalSparsity
from src.sparsity.empirical.empirical_validation import EmpiricalValidation
```

## 1. Initialize the Empirical Sparsity Framework

```python
framework = EmpiricalSparsity()
print("Framework loaded successfully!")
```

## 2. Load and Validate Raw Mutation Counts

We use the verified `mutation_counts.csv` located at `data/raw/empirical/`.

```python
counts_path = "data/raw/empirical/mutation_counts.csv"
counts_df = framework.load_counts(counts_path)
counts_df
```

## 3. Calculate Empirical Sparsity

We calculate frequencies, normalized frequencies, and the empirical sparsity score.

```python
results_df = framework.calculate(counts_df)
results_df
```

## 4. Save Outputs and Generate Artifacts

Save the calculated dataset to `results/empirical/empirical_sparsity.parquet` along with metadata and plots.

```python
output_dir = "results/empirical/"
parquet_path, summary_path, plot_path = framework.generate_outputs(results_df, output_dir)

print(f"Saved Core Parquet to: {parquet_path}")
print(f"Saved Summary CSV to:  {summary_path}")
print(f"Saved Plot to:         {plot_path}")
```

## 5. Load and Display Summary Statistics

```python
summary_df = pd.read_csv(summary_path)
summary_df
```

## 6. Run Scientific Validation

Verify that the calculated dataset satisfies all structural schemas and scientific properties.

```python
validator = EmpiricalValidation()
certified = validator.validate_dataset(parquet_path)
print(f"Dataset Certified: {certified}")
```
